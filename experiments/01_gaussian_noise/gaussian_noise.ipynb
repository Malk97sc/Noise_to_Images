# Image Degradation by Gaussian Noise

```python
import sys
sys.path.append('../')

import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

from utils import create_img, plot_img
```

```python
img = create_img()
plot_img(img)
```

## Creating the Gaussian Noise

```python
img = img.astype(np.float32) / 255.0
```

```python
x, y = img.shape
mean = 0
var = 0.01
sigma = np.sqrt(var)
gaussian = np.random.normal(loc=mean, scale=sigma, size=(x, y))
```

```python
gaussian = np.clip(gaussian, 0, 1)
plot_img(gaussian)
```

## Add a Gaussian Noise

To add Gaussian noise to an image, we use the following formula:

$$ g(x, y) = f(x, y) + n(x, y) $$

Where $ f(x,y) $ is the original image and $ n(x,y) $ is the Gaussian Noise.

```python
g = img + gaussian
```

```python
g = np.clip(g, 0, 1)
plot_img(g)
```
